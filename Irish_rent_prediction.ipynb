```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('irish_rent_full.csv')
df.head() # here e have to predict the rent_euro
```

```python
df.describe()
```

```python
df.info()
```

```python
df['bedrooms_num'].fillna(df['bedrooms_num'].mean(),inplace=True)#let's perform one hot encoding
```

```python
df['Period'] = df['half_year'].apply(lambda x:int(x[:4] + x[5]))
df.head()
```

```python
df.drop(columns=['half_year'],inplace=True)
```

```python
X = df.drop(columns=['rent_euro'],axis=1)
y = df['rent_euro']
```

```python
X
```

```python
X_one_encode = pd.get_dummies(X,columns=['county','province', 'area', 'location', 'property_type', 'bedrooms','Period'],drop_first=True)
X_one_encode
 
```

```python
df.columns
```

```python
X_one_encode = X_one_encode.astype(int)
X_one_encode
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.naive_bayes import GaussianNB
```

```python
# Target variable
y = df['rent_euro']

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X_one_encode,
    y,
    test_size=0.2,
    random_state=42
)
```

```python
scaler = StandardScaler()#It is necessary for knn,decison
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
#here only linearRegression and Randomforest can be performed as we have to predict continous varible that is rent
```

```python
from sklearn.linear_model import LinearRegression

# Initialize model
lr = LinearRegression()

# Train model
lr.fit(X_train_scaled, y_train)
```

```python
y_train_pred = lr.predict(X_train_scaled)
y_test_pred = lr.predict(X_test_scaled)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

# Training metrics
print("Training Performance")
print("MAE:", mean_absolute_error(y_train, y_train_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_train, y_train_pred)))
print("R2 Score:", r2_score(y_train, y_train_pred))

print("\nTesting Performance")
print("MAE:", mean_absolute_error(y_test, y_test_pred))
print("RMSE:", np.sqrt(mean_squared_error(y_test, y_test_pred)))
print("R2 Score:", r2_score(y_test, y_test_pred))
```

```python
import joblib

joblib.dump(lr, 'model.pkl')
joblib.dump(scaler, 'scaler.pkl')
joblib.dump(X.columns.tolist(), 'columns.pkl')
```
